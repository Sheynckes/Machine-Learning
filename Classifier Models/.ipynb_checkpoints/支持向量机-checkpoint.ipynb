# 基础理论——最大化分类间隔

支持向量机（Support Vector Machine，SVM）是另一种性能强大且广泛应用的学习算法，它可以看作对感知器的扩展。感知器算法的优化目标是最小化分类误差，而SVM的优化目标是最大化分类间隔。这里的间隔是指两个分离的超平面（决策边界）间的距离，而最靠近超平面的训练样本称作**支持向量**（Support Vector），如下图所示：

```python
%%html
<img src="img/SVM schematic diagram.png">
```

决策边界间具有较大的间隔意味着模型具有较小的泛化误差，而较小的间隔则意味着模型可能过拟合。为了对间隔最大化有个直观的认识，我们仔细观察一下两条平行的决策边界，我们分别称其为正、负超平面，可表示为：

$$
\begin{cases}
w_0 + \mathbf{w}^T \mathbf{x}_{pos} = 1\\
w_0 + \mathbf{w}^T \mathbf{x}_{neg} = -1
\end{cases}
$$

将这两个等式相减，可以得到：

$$\mathbf{w}^T (\mathbf{x}_{pos} - \mathbf{x}_{neg}) = 2$$

进一步，可以通过向量w的长度来对其进行**规范化**，由此得到如下等式，其左侧可以解释为正、负超平面间的距离，也就是我们要最大化的间隔：

$$\frac{\mathbf{w}^T (\mathbf{x}_{pos} - \mathbf{x}_{neg})}{||w||} = \frac{2}{||w||}$$

在样本被正确划分的前提下，最大化分类间隔就是使上述等式右侧最大化（等价于对$\frac{||w||}{2}$进行最小化处理），这也是SVM的**目标函数**，可以记为：

$$
\begin{cases}
w_0 + \mathbf{w}^T x^{(i)} \ge 1 & if \ y^{(i)} = 1\\
w_0 + \mathbf{w}^T x^{(i)} \lt -1 & if \ y^{(i)} = -1
\end{cases}
$$

这两个方程可以解释为：所有的负样本都落在负超平面一侧，而所有的正样本则在正超平面划分出的区域中。它们可以写成统一的形式：

$$y^{(i)} (w_0 + \mathbf{w}^T \mathbf{x}^{(i)}) \ge 1 \qquad \forall \ i$$

# 使用SVM解决非线性问题(1)：松弛变量

松弛变量$\xi$由Vladimir Vapnik在1995年引入，并由此提出所谓的软间隔分类。引入松弛变量$\xi$的目的在于：对于非线性可分的数据来说，需要放松线性约束条件，以保证在适当的罚项成本下，对错误分类的情况进行优化时能够收敛。

取值为正的松弛变量可以简单地加入到线性约束中：

$$
\begin{cases}
\mathbf{w}^T x^{(i)} \ge 1 & if \ y^{(i)} = 1 - \xi^{(i)}\\
\mathbf{w}^T x^{(i)} \lt -1 & if \ y^{(i)} = 1 + \xi^{(i)}
\end{cases}
$$

由此，新的优化目标为对下式进行最小化处理：

$$\frac{1}{2}||\mathbf{w}||^2 + C(\sum_{i}{\xi^{(i)}})$$

通过变量C，我们可以控制对错误分类的惩罚程度。C值较大时，对应的错误惩罚也较大；当选择小的C值时，则对错误分类的惩罚没那么严格。因此，可以使用参数C来控制间隔的大小，进而可以在模型的偏差（Bias）和方差（Variance）之间取得平衡。如下图所示，增加C值就意味着增加偏差，并降低方差。

# 使用SVM解决非线性问题(2)：核函数

支持向量机在机器学习爱好者中广受欢迎的另一原因是：它可以很容易地使用“核技巧”来解决非线性可分问题。核方法处理非线性可分数据的基本理念就是：通过映射函数$\phi(\cdot)$将样本的原始特征映射到一个使样本线性可分的更高维空间中。如下图所示，我们可以将二维数据集通过下列映射转换到新的三维特征空间中，从而使得样本线性可分：

$$\phi(x_1, x_2) = (z_1, z_2, z_3) = (x_1, x_2, x_1^2 + x_2^2)$$

由下图可见，对样本进行映射后，我们可以将图中的两个类别通过线性超平面进行分割。如果我们把此超平面映射回原始特征空间，则可线性分割两类数据的超平面就变为非线性的了。

```python
%%html
<img src="img/SVM_kernel.png">
```

映射函数理论上可行，但在构建新的特征空间带来非常大的计算成本（因为维度增加）。这是就要用到**核函数**的方法。在实践中，我们需要做的就是将点积$(x^{(i)})^T \cdot x^{(j)}$映射为$\phi(x^{(i)})^T \cdot \phi(x^{(j)})$。为了降低点积精确计算阶段的成本耗费，我们定义一个所谓的核函数：

$$k(x^{(i)}, x^{(j)}) = \phi(x^{(i)})^T \cdot \phi(x^{(j)})$$

使用最为广泛的核函数就是**径向基函数核**（Radial Basis Function kernel，**RBF** kernel）或**高斯核**（Gaussian kernel），其中$\gamma = \frac{1}{2\sigma^2}$是待优化的自由参数：

$$k(x^{(i)}, x^{(j)}) = exp(-\frac{||x^{(i)} - x^{(j)}||^2}{2\sigma^2}) = exp(-\gamma ||x^{(i)} - x^{(j)}||^2)$$

粗略而言，“核”可解释为两个样本之间的“相似函数”。此处的负号将距离转换为相似性评分，而由于指数项的存在，使得相似性评分会介于0（样本差异巨大）和1（样本完全相同）之间。

# Coding

```python
# 导入需要的库

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

## 辅助函数

借助以下`plot_decision_region()`函数，可以绘制模型的决策区域，并观察不同样本的分类效果；其中，使用灰色小圆圈显示来自测试数据集的样本。

```python
# %load scripts/plot_decision_region.py
# @Time    : 2018/5/10 22:45
# @Author  : SHeynckes
# @Email   : sheynckes@outlook.com
# @File    : plot_decision_region.py
# @Software: PyCharm


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    # setup marker generator and color map
    markers = ("s", "x", "o", "^", "v")
    colors = ("red", "blue", "lightgreen", "gray", "cyan")
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min = X[:, 0].min() - 1
    x1_max = X[:, 0].max() + 1
    x2_min = X[:, 1].min() - 1
    x2_max = X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot all samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8,
                    c=cmap(idx), marker=markers[idx], label=cl)

    # highlight test samples
    if test_idx:
        X_test = X[test_idx, :]
        y_test = y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1], c="grey", alpha=1.0,
                    linewidth=1, marker="o", s=55, label="Test Set")
```

## 使用SVM模型处理Iris数据集

首先，通过`scikit-learn`中的`datasets`直接导入Iris数据集。提取150个花朵样本中的花瓣长度（petal length）和花瓣宽度（petal width）两个特征的值，并由此构建特征矩阵X，同时将对应花朵所属类型的类标赋值给向量y。

同时，为了评估训练得到的模型在未知数据上的表现，需要将数据集划分为训练数据集(含有105个数据）和测试数据集（含有45个数据）。

```python
iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

使用`scikit-learn`的`preprocessing`模块中的`StandardScaler`类对特征进行标准化处理。使用`StandardScaler`中的`fit()`方法，可以计算训练数据中每个特征的$\mu$（样本均值）和$\sigma$（标准差）。通过调用`transform()`方法，可以使用前面计算得到的$\mu$和$\sigma$来对训练数据做标准化处理。需注意的是，我们要使用相同的缩放参数分别处理训练和测试数据集，以保证它们的值是彼此相当的。

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

接下来，就可以对SVM模型进行训练了。该模型中的参数C是惩罚项（松弛变量和）的系数。

这里，比较三种SVM模型的计算结果，分别计算每种模型的错分类数量、分类准确率，并利用`plot_decision_regions()`函数绘制模型的决策区域。

```python
# 1. 选择线性核函数构建SVM模型

svm_linear = SVC(kernel="linear", C=1.0, random_state=0)
svm_linear.fit(X_train_std, y_train)
y_pred = svm_linear.predict(X_test_std)
# 计算错分类数量
print("Misclassified samples: %d" % (y_test != y_pred).sum())
# 计算分类准确率
print("Accuracy: %.2f" % accuracy_score(y_test, y_pred))
# 绘制决策区域
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X=X_combined_std, y=y_combined, 
                      classifier=svm_linear, test_idx=range(105, 150))
plt.xlabel("petal length [standardized]")
plt.ylabel("petal width [standardized]")
plt.title("Linear Kernel Function")
plt.legend(loc="upper left")
plt.show()
```

```python
# 2. 选择RBF核函数构建SVM模型(gamma=0.2)

svm_rbf_small_gamma = SVC(kernel="rbf", C=1.0, gamma=0.2, random_state=0)
svm_rbf_small_gamma.fit(X_train_std, y_train)
y_pred = svm_rbf_small_gamma.predict(X_test_std)
# 计算错分类数量
print("Misclassified samples: %d" % (y_test != y_pred).sum())
# 计算分类准确率
print("Accuracy: %.2f" % accuracy_score(y_test, y_pred))
# 绘制决策区域
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X=X_combined_std, y=y_combined, 
                      classifier=svm_rbf_small_gamma, test_idx=range(105, 150))
plt.xlabel("petal length [standardized]")
plt.ylabel("petal width [standardized]")
plt.title("RBF Kernel Function (gamma=0.2)")
plt.legend(loc="upper left")
plt.show()
```

```python
# 3. 选择RBF核函数构建SVM模型(gamma=100.0)

svm_rbf_large_gamma = SVC(kernel="rbf", C=1.0, gamma=100.0, random_state=0)
svm_rbf_large_gamma.fit(X_train_std, y_train)
y_pred = svm_rbf_large_gamma.predict(X_test_std)
# 计算错分类数量
print("Misclassified samples: %d" % (y_test != y_pred).sum())
# 计算分类准确率
print("Accuracy: %.2f" % accuracy_score(y_test, y_pred))
# 绘制决策区域
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X=X_combined_std, y=y_combined, 
                      classifier=svm_rbf_large_gamma, test_idx=range(105, 150))
plt.xlabel("petal length [standardized]")
plt.ylabel("petal width [standardized]")
plt.title("RBF Kernel Function (gamma=100.0)")
plt.legend(loc="upper left")
plt.show()
```

## 使用SVM处理『异或』数据集

使用`NumPy`库中的`logical_xor()`函数创建了一个经过『异或』操作的数据集，其中100个样本属于类别1，另外的100个样本被划定为类别-1。使用`scikit-learn`训练一个SVM分类器（使用rbf核）, 使之可以通过一个非线性决策边界来对该『异或』数据进行分类。

创建了一个『异或』数据集，并绘制其散点图。

```python
np.random.seed(0)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0]>0, X_xor[:, 1]>0)
y_xor = np.where(y_xor, 1, -1)

plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], c="b", marker="x", label="1")
plt.scatter(X_xor[y_xor==-1, 0], X_xor[y_xor==-1, 1], c="r", marker="s", label="-1")
plt.ylim(-3.0)
plt.legend()
plt.show()
```

使用RBF核函数构建SVM分类模型（`gamma=0.10， C=10.0`）。由于这里使用的gamma值较小，会增加受影响的训练样本的范围，导致决策边界更加宽松。

```python
svm = SVC(kernel="rbf", random_state=0, gamma=0.10, C=10.0)
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc="upper left")
plt.show()
```
